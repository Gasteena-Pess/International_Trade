# STEP 2 : DOMAIN KNOWLEDGE

### Key Columns:
Period: Likely the time (e.g., year, month, or quarter) during which the trade data was recorded.

Balance Total: The total trade balance, which is the difference between total exports and imports. A positive balance indicates a trade surplus, while a negative balance indicates a deficit.

Balance Goods BOP: Trade balance specifically for goods, based on the Balance of Payments (BOP) system.

Balance Services: Trade balance specifically for services.

Exports Total: Total exports, including both goods and services.

Exports Goods BOP: Exports of goods only, based on the BOP system.

Exports Services: Exports of services.

Imports Total: Total imports, including both goods and services.

Imports Goods BOP: Imports of goods only, based on the BOP system.

Imports Services: Imports of services.

### Why is it important?
Economic Health Indicator: Trade data serves as a crucial indicator of a countryâ€™s economic health. A positive trade balance (more exports than imports) can indicate economic strength, while a negative balance may suggest economic vulnerabilities.

Policy Formulation: Understanding trade patterns assists policymakers in developing effective trade policies. It helps in analyzing the impacts of tariffs, trade agreements, and international relationships.

Market Competitiveness: Businesses and analysts can assess the U.S. competitiveness in global markets by evaluating export and import trends. This data can inform strategic decisions for businesses looking to expand internationally or adjust to changing market conditions.

Investment Decisions: Investors often look at trade balances and trends to gauge market potential and make informed investment decisions in sectors affected by international trade.

Global Supply Chain Insights: The dataset can provide insights into supply chain dynamics, helping businesses and governments understand dependencies on foreign goods and services.

Trend Analysis: Historical trade data allows for trend analysis over time, facilitating long-term economic forecasts and strategic planning.

# STEP 3 : DATA CLEANING

```python
# Handle Missing Values
print("\nMissing Values in each column:\n", df.isnull().sum())

# Dropping rows with all missing values
df.dropna(how='all', inplace=True)
```

```python
print("\nNumber of duplicate rows:", df.duplicated().sum())
df.drop_duplicates(inplace=True)
```

```python
for column in df.columns:
    print(f"{column} unique values: {df[column].unique()}")
```

```python
df.columns = df.columns.str.strip()
# Convert 'Period' to datetime
df['Period'] = pd.to_datetime(df['Period'], errors='coerce')

# Convert relevant object columns to numeric
numeric_columns = [
    'Balance Total',
    'Balance Goods BOP',
    'Balance Services',
    'Exports Total',
    'Exports Goods BOP',
    'Exports Services',
    'Imports Total',
    'Imports Goods BOP',
    'Imports Services'
]

for col in numeric_columns:
    df[col] = pd.to_numeric(df[col].astype(str).str.replace(',', '').replace('nan', '0'), errors='coerce')

print("\nUpdated Data Types:")
print(df.dtypes)

print("\nUpdated DataFrame:")
print(df.head())
```

```python
#  Basic statistics summary of the dataset 
print("\nStatistical Summary of the dataset:\n")
print(df.describe(include='all'))
```

# STEP 4 : DATA ANALYSIS & STEP 5 : DATA VISUALIZING

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Descriptive Statistics
print("Descriptive Statistics:")
print(df.describe())
```

Key statistics:

Mean Balance Total: -187,736 (indicating an average trade deficit over time).

Mean Exports Total: 689,256

Mean Imports Total: 876,992

Trade Deficit: On average, the country has a negative balance, indicating persistent trade deficits.

Exports and Imports: Both have been growing, with imports slightly outpacing exports, leading to trade deficits.

```python
# Correlation Analysis
correlation_matrix = df.corr()
print("\nCorrelation Matrix:")
print(correlation_matrix)
```

```python
# Visualizing Correlation Matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True)
plt.title('Correlation Matrix')
plt.show()
```

Imports and Exports have a strong positive correlation (0.99), suggesting that higher exports often coincide with higher imports.

Trade balance has a strong negative correlation with imports and exports, indicating that as imports grow, the trade balance worsens.

This graph tells us which variables are strongly related. Policymakers can see that controlling imports may directly improve the trade balance.

```python
#  Distribution Analysis
plt.figure(figsize=(15, 10))
for i, col in enumerate(['Balance Total', 'Exports Total', 'Imports Total']):
    plt.subplot(3, 1, i + 1)
    sns.histplot(df[col].dropna(), bins=30, kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

The Balance Total shows a left-skewed distribution, confirming that trade deficits are more common than surpluses.

Both exports and imports distributions show steady growth, with no extreme outliers.

Helps to understand the distribution of trade metrics over time. The left skewness in the balance confirms the economyâ€™s trade deficit tendencies.

```python
# Time Series Decomposition
from statsmodels.tsa.seasonal import seasonal_decompose
decomposition = seasonal_decompose(df['Balance Total'].dropna(), model='additive', period=12)
plt.figure(figsize=(12, 8))
decomposition.plot()
plt.show()
```

The Balance Total remains mostly negative over time, with some small peaks indicating trade surpluses.

Both exports and imports show an upward trend, with imports growing faster than exports, leading to increasing trade deficits over time.

This graph tracks the historical trends in trade. The upward trend in imports compared to exports signals areas of concern, such as over-reliance on imported goods.

```python

# Plotting Balance Total, Exports Total, and Imports Total over Time
plt.plot(df['Period'], df['Balance Total'], label='Balance Total', color='b')
plt.plot(df['Period'], df['Exports Total'], label='Exports Total', color='g')
plt.plot(df['Period'], df['Imports Total'], label='Imports Total', color='r')

plt.title('Trends in Trade Values Over Time')
plt.xlabel('Year')
plt.ylabel('Trade Values (in millions)')
plt.legend()
plt.grid()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

The trend shows a clear downward trajectory, indicating worsening trade balances.

The seasonal component shows periodic fluctuations, which may be tied to global trade cycles, holidays, or fiscal policies.

By decomposing the trade balance, we can separate long-term trends from seasonal fluctuations, helping analysts to better forecast and understand trade dynamics.

```python
# Visualizing Exports vs Imports
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Imports Total', y='Exports Total', data=df)
plt.title('Exports vs Imports')
plt.xlabel('Imports Total (in millions)')
plt.ylabel('Exports Total (in millions)')
plt.grid()
plt.show()
```

There is a clear linear relationship between exports and imports, meaning countries that export more also tend to import more.

No significant outliers show extreme imbalances between exports and imports.

This helps to visualize the relationship between exports and imports, confirming their close relationship. Policymakers can use this to find opportunities for trade policy adjustments.

```python
#Box Plots for Outlier Detection
plt.figure(figsize=(15, 8))
sns.boxplot(data=df[['Balance Total', 'Exports Total', 'Imports Total']].dropna())
plt.title('Box Plots for Trade Values')
plt.ylabel('Trade Values (in millions)')
plt.xticks(rotation=45)
plt.show()
```

Balance Total shows significant outliers, likely from extreme deficits in certain years.

Exports and Imports are more stable, with fewer extreme outliers.

Box plots help to detect outliers that might distort overall analysis. In this case, itâ€™s clear that some years have experienced exceptionally high trade deficits.
